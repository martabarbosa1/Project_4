```python
#Choose books
drop_box = driver.find_element('id', 'searchDropdownBox')
drop = 'Books'
# send_keys() to simulate key strokes
drop_box.send_keys(drop)
# # locate submit button by_class_name
drop_button = driver.find_element('id', "nav-search-submit-button")  #TODO: fix. Says the button is not there.
# # .click() to mimic button click
drop_button.click()

# book_names = ['Atomic Habits', 'The Light We Carry', 'I\'m Glad My Mom Died', '1234567tt', 'pão de açúcar']
        
def retrieve_price(x):
    search_box = driver.find_element('id', 'twotabsearchtextbox')
    #erase everything and then send keys
    search_box.send_keys(Keys.CONTROL, 'a')
    search_box.send_keys(Keys.DELETE)
    #send_keys() to simulate key strokes
    search_box.send_keys(x)
    search_button = driver.find_element('id', "nav-search-submit-button")  #TODO: fix. Says the button is not there.
    search_button.click()
 
    
    #Click first appearance (usually the best seller)
    try:
        group = driver.find_element(By.XPATH, "//*[@id=\"search\"]/div[1]/div[1]/div/span[1]/div[1]/div[2]/div/div/div/div/div/div[1]/div/div[2]/div/span/a/div/img")
        group.click()
    except NoSuchElementException:
        #Choose books
        drop_box = driver.find_element('id', 'searchDropdownBox')
        drop = 'Books'
        # send_keys() to simulate key strokes
        drop_box.send_keys(drop)
        # # locate submit button by_class_name
        drop_button = driver.find_element('id', "nav-search-submit-button")  #TODO: fix. Says the button is not there.
        # # .click() to mimic button click
        drop_button.click()
        
        print("exception handled")

    #parse the page using beautiful soup library
    soup_actual = BeautifulSoup(driver.page_source, 'html.parser')

    #grab the tag for the amazon best suggested price
    items = soup_actual.find_all('li', {"class": "swatchElement"})


    items_text = list(map(lambda a : a.text, items))
        
    return items_text, x

# result = list(map(retrieve_price, book_names))
# result
```
